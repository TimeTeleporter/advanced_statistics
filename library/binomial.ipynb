# Binomial models
## Globe tossing experiment

We model a globe tossing experiment to figure out the share of water on the earths surface. The experiment yielded 13 water out of 24 tosses.

```r
library(rethinking)
d <- data.frame(list(water = c(13), tosses = c(24)))
d
```

```r
# Binomial generalized linear model
model <- ulam(
	alist(
		water ~ dbinom(tosses, p),
		logit(p) <- a,
		a ~ dnorm(0, 100)
	),
	data = d, chains = 8, cores = 8#, log_lik = TRUE
)
```

```r
post <- extract.samples(model, n = 10000)
p <- logistic(post$a)

frac_seq <- seq(0, 1, 0.01)
```
